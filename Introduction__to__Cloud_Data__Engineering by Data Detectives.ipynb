```python
# 2. Import Libraries
import pandas as pd
import numpy as np
# 3. Load Dataset
df = pd.read_csv('/content/drive/MyDrive/processed_data.csv')
df.head()
df
```

```python
# Check for missing values
df.isnull().sum()

```

```python
# 4. Data Summary
print("Shape:", df.shape)
print("\nNull Values:\n", df.isnull().sum())
print("\nData Types:\n", df.dtypes)
```

```python
# Convert 'Order Date' to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.head()
```

```python
# Convert 'Year-Month' to datetime
df['Year-Month'] = pd.to_datetime(df['Year-Month'], format='%Y-%m')
df.head()
```

```python
# Create 'Total Revenue' = Amount * Quantity
df['Total Revenue'] = df['Amount'] * df['Quantity']
df.head()
```

```python
# Extract year from 'Order Date'
df['Order Year'] = df['Order Date'].dt.year
df.head()
```

```python
# Reorder columns
columns_order = ['Order ID', 'Order Date', 'Order Year', 'Year-Month', 'CustomerName', 'City',
                 'State', 'Category', 'Sub-Category', 'Quantity', 'Amount', 'Total Revenue',
                 'Profit', 'PaymentMode']
df = df[columns_order]
df.head()
```

```python
# Save final cleaned dataset
cleaned_file_path = '/content/drive/My Drive/cleaned_sales_data.csv'
df.to_csv(cleaned_file_path, index=False)
print("✅ Cleaned dataset saved to Google Drive.")
```
